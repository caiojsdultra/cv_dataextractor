```python
# Importando bibliotecas
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pytesseract
```

```python
# Instanciando o Tesseract-OCR
# Antes de tudo, é necessário que você instale o Tesseract-OCR e depois aponte o executável conforme descrito abaixo:
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
# Vamos dar uma olhada na imagem que trabalharemos. Observe todas as variáveis: Imagem colorida, resolução não muito boa, frase a ser extraída em branco etc
picture = np.array(Image.open(r'C:\Users\*\Documents\Job\Pictures\catmeme.jpg'))
plt.figure(figsize=(10,10))
plt.title('CAT MEME')
plt.imshow(picture); plt.xticks([]); plt.yticks([])
plt.savefig('catimg1.png')
```

```python
# Utilizaremos a função bilateralFilter() para retirar o máximo de ruídos possíveis da nossa imagem.

picture= cv2.bilateralFilter(picture,5, 55,60)
plt.figure(figsize=(10,10))
plt.title('BILATERAL FILTER')
plt.imshow(picture); plt.xticks([]); plt.yticks([])
plt.savefig('catimg2.png',bbox_inches='tight')
```

```python
# Vamos aplicar uma escala cinza nessa imagem para facilitar o thresholding.

picture = cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.title('CAT GRAYSCALE PIC')
plt.imshow(picture, cmap='gray'); plt.xticks([]); plt.yticks([])
plt.savefig('catimage3.png',bbox_inches='tight')
```

```python
# Aplicando Threshold na imagem alvo para binarização
# Link: https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

_, picture = cv2.threshold(picture,240, 255, 1)
    
plt.figure(figsize=(10,10))
plt.title('BINARY CAT MEME')
plt.imshow(picture, cmap='gray'); plt.xticks([]); plt.yticks([])
plt.savefig('catimage4.png',bbox_inches='tight')
```
